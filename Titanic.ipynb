```python
### FILLING MISSING VALUES

train['Age'].fillna(value=train.Age.mean(), inplace=True)
test['Age'].fillna(value=train.Age.mean(), inplace=True)
train['Embarked'].fillna(value='S', inplace=True)
test['Fare'].fillna(value=test['Fare'].mean(), inplace=True)
```

```python
### CONVERTING CONTINUOUS 'Age' INTO CATEGORICAL VARIABLE 'Agebin'

binwidth=int((train.Age.max()-train.Age.min())/5)
bins=range(int(train.Age.min()),int(train.Age.max()),binwidth)
group=[1,2,3,4,5]
train['Agebin']= pd.cut(train['Age'],bins,labels=group, include_lowest=True)





binwidth=int((test.Age.max()-test.Age.min())/5)
bins=range(int(test.Age.min()),int(test.Age.max()),binwidth)
group=[1,2,3,4,5]
test['Agebin']= pd.cut(test['Age'],bins,labels=group, include_lowest=True)



test['Agebin'].describe()
```

```python
### ### CONVERTING CONTINUOUS 'Fare' INTO CATEGORICAL VARIABLE 'Range'



binwidth=int((train.Fare.max()-train.Fare.min())/5)
bins=range(int(train.Fare.min()),int(train.Fare.max()),binwidth)
group=[1,2,3,4,5]
train['Range']= pd.cut(train['Fare'],bins,labels=group, include_lowest=True)

binwidth=int((test.Fare.max()-test.Fare.min())/5)
bins=range(int(test.Fare.min()),int(test.Fare.max()),binwidth)
group=[1,2,3,4,5]
test['Range']= pd.cut(test['Fare'],bins,labels=group, include_lowest=True)

test['Range'].describe()
```

```python
### FILLING MISSING VALUES OF AGEBIN AND RANGE:
train['Agebin'].fillna(value=5, inplace=True)
train['Range'].fillna(value=5, inplace=True)

test['Agebin'].fillna(value=5, inplace=True)
test['Range'].fillna(value=5, inplace=True)

```

```python
### PREPARING X AND X_TEST:

y=train['Survived'].values

X=train[['Pclass','Sex','Embarked','Agebin','Range','Family']].values
X_test=test[['Pclass','Sex','Embarked','Agebin','Range','Family']].values
```

```python
### CONVERTING STRING CATEGORICAL VARIABLES TO NUMERIC VARIABLES


lesex=preprocessing.LabelEncoder()
le_embarked=preprocessing.LabelEncoder()
le_age=preprocessing.LabelEncoder()
le_range=preprocessing.LabelEncoder()


## FOR SEX

lesex.fit(['female','male'])

X[:,1]=lesex.transform(X[:,1])

X_test[:,1]=lesex.transform(X_test[:,1])

## FOR EMBARKED

le_embarked.fit(['S','Q','C'])

X[:,2]=le_embarked.transform(X[:,2])

X_test[:,2]=le_embarked.transform(X_test[:,2])




```

```python
### X_ AND X_test_ ARE WITHOUT STANDARDISED 


X_=X.copy()
X_test_=X_test.copy()
```

```python
### STANDARDISING X AND X_TEST:


X=preprocessing.StandardScaler().fit(X).transform(X)
X_test=preprocessing.StandardScaler().fit(X_test).transform(X_test)
```

```python
###TEST TRAIN SPLIT:

X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)

###WITHOUT STANDARDISED:

x_train,x_test_,Y_train,Y_test=train_test_split(X_,y,test_size=0.3,random_state=0) 
```

```python
### LOGISTIC REGRESSION:(WITH STANDARDISED FEATURES)

lr=LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)

yhat=lr.predict(x_test)

acc(y_test,yhat)*100

```

```python
### CROSS VALIDATION

scores=cross_val_score(lr,X_train,y_train,cv=5)

yhat0=cross_val_predict(lr,X_train,y_train,cv=5)

np.mean(scores)*100
```

```python
### KNN:(WITHOUT STANDARDISED FEATURES)

from sklearn.neighbors import KNeighborsClassifier

k=5

neigh=KNeighborsClassifier(n_neighbors=k).fit(x_train,Y_train)

yhat1=neigh.predict(x_test_)

acc(Y_test,yhat1)*100
```

```python
### DECISION TREE:(WITHOUT STANDARDISED FEATURES)

from sklearn.tree import DecisionTreeClassifier

tree=DecisionTreeClassifier(criterion='entropy', max_depth=5)



tree.fit(x_train,Y_train)

yhat2=tree.predict(x_test_)

acc(Y_test,yhat2)*100
```

```python
### SVM (LINEAR)(STANDARDISED FEATURES)

from sklearn import svm

clf1=svm.SVC(kernel='linear', C=6)

clf1.fit(X_train,y_train)

yhat3=clf1.predict(x_test)

(acc(y_test,yhat3))*100
```

```python
### SVM (POLYNOMIAL)(STANDARDISED FEATURES)

from sklearn import svm

clf2=svm.SVC(kernel='poly',degree=4,coef0=1, C=6)

clf2.fit(X_train,y_train)

yhat3=clf2.predict(x_test)

(acc(y_test,yhat3))*100
```

```python
### SVM (RBF)(STANDARDISED FEATURES)

from sklearn import svm

clf3=svm.SVC(kernel='rbf',gamma=2, C=7)

clf3.fit(X_train,y_train)

yhat3=clf3.predict(x_test)

(acc(y_test,yhat3))*100
```

```python
### SVM REGRESSION (LINEAR)(STANDARDISED FEATURES)

from sklearn.svm import LinearSVR

svm_reg=LinearSVR(epsilon=1.5)

svm_reg.fit(X_train,y_train)

yhat4=svm_reg.predict(x_test)

(acc(y_test,yhat4))*100
```

```python
### VOTING CLASSIFIER(ENSEMBLE)(WITHOUT STANDARDISED FEATURES)

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC

lr=LogisticRegression()
rnd=RandomForestClassifier()
svm=SVC()
dt=DecisionTreeClassifier()
knn=KNeighborsClassifier()

voting_clf=VotingClassifier(estimators=[('lr',lr),('rf', rnd),('dt',dt),('knn',knn)], voting='hard')

voting_clf.fit(x_train,Y_train)

yhat6=voting_clf.predict(x_test_)

(acc(Y_test,yhat6))*100
```

```python
### RANDOM FOREST CLASSIFIER

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, accuracy_score
from sklearn.model_selection import GridSearchCV

# Choose the type of classifier. 
rnd = RandomForestClassifier()


### TO GET PARAMETERS OF ANY MODEL:

rnd.get_params().keys()

```

```python
### GRIDSEARCH + RANDOM FOREST CLASSIFIER:


# Choose some parameter combinations to try
parameters = {'n_estimators': [4, 6, 9], 
              'max_features': ['log2', 'sqrt','auto'], 
              'criterion': ['entropy', 'gini'],
              'max_depth': [2, 3, 5, 10], 
              'min_samples_split': [2, 3, 5],
              'min_samples_leaf': [1,5,8]
             }

# Type of scoring used to compare parameter combinations
acc_scorer = make_scorer(accuracy_score)

# Run the grid search
grid_obj = GridSearchCV(rnd, parameters, scoring=acc_scorer)
grid_obj = grid_obj.fit(x_train, Y_train)

# Set the clf to the best combination of parameters
rnd = grid_obj.best_estimator_

# Fit the best algorithm to the data. 
rnd.fit(x_train, Y_train)

yhat7=rnd.predict(x_test_)

(acc(Y_test,yhat7))*100
```

```python
### TRAINING ON FULL GIVEN DATASET

rnd.fit(X,y)


yhat_test=rnd.predict(X_test)

yhat_test
```

```python
### PREPARING FOR KAGGLE SUBMISSION FORMAT

df = pd.DataFrame(yhat_test)
df.columns=['Survived']

pred = pd.concat([test, df], axis = 1)

pred.index=pred.PassengerId
pred=pred[['Survived']]

pred.to_csv('prediction.csv')

```
